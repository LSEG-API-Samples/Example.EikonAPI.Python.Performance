Chart how dividend payout effects the stock price
===

```python
import eikon as ek
import pandas as pd
import plotly.offline as offline
import plotly.graph_objs as go
```

```python
offline.offline.init_notebook_mode(connected=True)
```

```python
ek.set_app_id('your_eikon_app_id')
```

```python
msft = ek.get_timeseries(["MSFT.O"], fields='CLOSE', start_date="2015-01-01", end_date="2017-01-01")
msft.head()
```

```python
fields = ['TR.DivExDate', 'TR.DivAdjustedGross', 'TR.DivAdjustedNet']
div,err = ek.get_data('MSFT.O', fields=fields, parameters={'SDate':'01Jan15', 'EDate':'01Jan17', 'DateType':'ED'})
div
```

```python
msft_price = go.Scatter(
    x=msft.index,
    y=msft['CLOSE'],
    name='Close Price'
);

msft_div = go.Scatter(
    x=div['Dividend Ex Date'],
    y=[msft['CLOSE'][x] for x in div['Dividend Ex Date'].tolist() ],
    text=div['Dividend Ex Date'] + '<br><b>Gross:</b> ' + div['Adjusted Gross Dividend Amount'].astype(str) + '<br><b>Net:</b> ' + div['Adjusted Net Dividend Amount'].astype(str),
    name='',
    mode='markers',
    marker = dict(size = 10),    
    hoverinfo='text'
);

layout = go.Layout(
    showlegend=False,
    title = 'Microsoft Dividends'
);

fig = go.Figure(data=[msft_price, msft_div], layout=layout)
offline.iplot(fig)
```
